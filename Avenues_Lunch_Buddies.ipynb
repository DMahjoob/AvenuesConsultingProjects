```python
import pandas as pd
import io
# comment here as a tester
df = pd.read_csv("Responses.csv")
df = df.drop(['Timestamp'], axis=1)
```

```python
memberList = []
num_rows = len(df)
row_index = 0

while row_index < num_rows:
  memberList.append([', '.join(map(str, df.iloc[row_index]))])
  row_index += 1

print(memberList)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import random

# Load the SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Initialize variables
taken_people = set()  # To track people who are already matched
matches = []  # List to store the final pairs

# Function to calculate the similarity of a specific field (e.g., favorite food)
def calculate_similarity(member, member2):
    # Example: Assuming the second column in each row is the "favorite food"
    field1 = member[0].split(',')[1].strip()
    field2 = member2[0].split(',')[1].strip()

    # Get sentence embeddings for each field
    embedding1 = model.encode([field1])
    embedding2 = model.encode([field2])

    # Calculate cosine similarity
    similarity_score = cosine_similarity(embedding1, embedding2)[0][0]
    return similarity_score

# Main matching loop
for member in memberList:
    person_name = member[0].split(',')[0]
    if person_name in taken_people:
        # This person is already matched, skip
        continue

    match_list = []
    # Compare this member to others
    for member2 in memberList:
        if member == member2:
            continue

        other_person = member2[0].split(',')[0]
        if other_person in taken_people:
            # This other person is already matched, skip
            continue

        # Calculate similarity and add a small random factor to break ties
        similarity_score = calculate_similarity(member, member2) + random.randint(-1,1)/10.0
        match_list.append((similarity_score, other_person))

    # Sort potential matches by similarity score descending
    sorted_matches = sorted(match_list, key=lambda x: x[0], reverse=True)

    # Pick the best match if available
    if sorted_matches:
        best_match_score, matched_person = sorted_matches[0]

        # Mark both as taken
        taken_people.add(person_name)
        taken_people.add(matched_person)

        # Store the match
        matches.append((person_name, matched_person, best_match_score))

# Identify any unmatched members
unmatched_members = [m for m in memberList if m[0].split(',')[0] not in taken_people]

# If there are leftover members, pair them off arbitrarily
if unmatched_members:
    for i in range(0, len(unmatched_members), 2):
        if i + 1 < len(unmatched_members):
            person1 = unmatched_members[i][0].split(',')[0]
            person2 = unmatched_members[i+1][0].split(',')[0]

            # Add these forced matches to the list
            matches.append((person1, person2, 0.0))

            # Mark them as taken
            taken_people.add(person1)
            taken_people.add(person2)

# Print the matches
for match in matches:
    print(f"Best Match: {match[0]} and {match[1]} with a similarity score of {match[2]:.4f}")

print(memberList)
```
